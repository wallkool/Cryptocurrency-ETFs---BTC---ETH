```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os 
excel_file = r'C:\Users\ADMIN\Project\ETF_Predict_After_ETF_Project\Price Data\All.xlsx'
```

```python
sheet_names = ['ARKB', 'BRRR', 'IBIT', 'BTCO','IBIT', 'BTC', 'ETH', 'FBTC', 'HODL', 'GBTC']

# Duyt qua tng sheet v c d liu
for sheet_name in sheet_names:
    # c d liu t mi sheet vo DataFrame
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    
    # Thc hin cc thao tc vi DataFrame  y
    # V d: Hin th biu  ct cho ct u tin v ct th hai (gi s l ct ngy v gi)
    plt.figure(figsize=(10, 5))
    plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=sheet_name)  # Thay i cho ph hp vi ct c th ca bn
    plt.title(f"Biu  cho sheet {sheet_name}")
    plt.xlabel('Ngy')  # Thay i cho ph hp vi tn ct ca bn
    plt.ylabel('Gi')  # Thay i cho ph hp vi tn ct ca bn
    plt.legend()
    plt.grid(True)
    
    # Lu biu  di dng hnh nh
    plt.savefig(f"{sheet_name}_chart.png")
    
    # Hin th biu 
    plt.show()
```
