```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os 
excel_file = r'C:\Users\ADMIN\Project\ETF_Predict_After_ETF_Project\Price Data\All.xlsx'
```

```python
sheet_names = ['ARKB', 'BRRR', 'IBIT', 'BTCO','IBIT', 'BTC', 'ETH', 'FBTC', 'HODL', 'GBTC']

# Duyệt qua từng sheet và đọc dữ liệu
for sheet_name in sheet_names:
    # Đọc dữ liệu từ mỗi sheet vào DataFrame
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    
    # Thực hiện các thao tác với DataFrame ở đây
    # Ví dụ: Hiển thị biểu đồ cột cho cột đầu tiên và cột thứ hai (giả sử là cột ngày và giá)
    plt.figure(figsize=(10, 5))
    plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=sheet_name)  # Thay đổi cho phù hợp với cột cụ thể của bạn
    plt.title(f"Biểu đồ cho sheet {sheet_name}")
    plt.xlabel('Ngày')  # Thay đổi cho phù hợp với tên cột của bạn
    plt.ylabel('Giá')  # Thay đổi cho phù hợp với tên cột của bạn
    plt.legend()
    plt.grid(True)
    
    # Lưu biểu đồ dưới dạng hình ảnh
    plt.savefig(f"{sheet_name}_chart.png")
    
    # Hiển thị biểu đồ
    plt.show()
```
